```python
import random
import numpy as np
import matplotlib.pyplot as plt
```

```python
dist_map = {"uniform":random.uniform, "poisson":np.random.poisson, "normal": random.gauss}
```

```python
def generate_randoms(sample, dist_map, required_distribution, a=0, b=1):
    dist_values = []
    
    if required_distribution=="uniform":
        for i in range(sample):
            dist_values.append(dist_map[required_distribution](a, b))
    else:
        for i in range(sample):
            dist_values.append(dist_map[required_distribution]())
    
    return dist_values
```

```python
# uniform
```

```python
uniform_50 = generate_randoms(50, dist_map, "uniform", 0, 1)
uniform_500 = generate_randoms(500, dist_map, "uniform", 0, 1)
uniform_5000 = generate_randoms(5000, dist_map, "uniform", 0, 1)
uniform_1L = generate_randoms(100000, dist_map, "uniform", 0, 1)
```

```python
# poisson
```

```python
poisson_50 = generate_randoms(50, dist_map, "poisson")
poisson_500 = generate_randoms(500, dist_map, "poisson")
poisson_5000 = generate_randoms(5000, dist_map, "poisson")
poisson_1L = generate_randoms(100000, dist_map, "poisson")
```

```python
# gaussian
```

```python
normal_50 = generate_randoms(50, dist_map, "normal")
normal_500 = generate_randoms(500, dist_map, "normal")
normal_5000 = generate_randoms(5000, dist_map, "normal")
normal_1L = generate_randoms(100000, dist_map, "normal")
```

```python
plt.hist(normal_5000, bins=50, edgecolor='black')
plt.show()
```
