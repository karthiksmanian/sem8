```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
df = pd.read_csv("Z:/20XWO8-DATA-VIS/ps6/company_sales_data.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
x_coords = np.array(df['month_number'])
y_coords = np.array(df['total_profit'])
```

```python
# 1
plt.plot(x_coords, y_coords, color='orange')
plt.xticks(x_coords)
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.show()
```

```python
# 2
x_coords = np.array(df['month_number'])
y_coords = np.array(df['total_units'])
plt.plot(x_coords, y_coords, color='red', linestyle='dotted', linewidth=3, marker= 'o')
plt.xlabel('Month Number')
plt.xticks(x_coords)
plt.ylabel('Total Units')
plt.show()
```

```python
# 3
x = np.array(df['month_number'])
y1 = np.array(df['facecream'])
y2 = np.array(df['facewash'])
y3 = np.array(df['toothpaste'])
y4 = np.array(df['bathingsoap'])
y5 = np.array(df['shampoo'])
y6 = np.array(df['moisturizer'])

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.plot(x, y4)
plt.plot(x, y5)
plt.plot(x, y6)
plt.xticks(x)
plt.show()
```

```python
# 4
plt.scatter(x, y3)
plt.xticks(x)
plt.show()
```

```python
# 5
plt.bar(['Facecream', 'Facewash'], [np.sum(y1), np.sum(y2)], width=0.2, align='center')
plt.show()
```

```python
# 6
plt.bar(x, y4, width=0.5, align='center')
plt.xticks(x)
plt.show()
```

```python
# 7

plt.hist(df['total_profit'], bins=6, edgecolor='black')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit for Each Month')
plt.show()
```

```python
# 8
res = []
for col in df[df.columns[1:-2]]:
    res.append(df[col].sum())

plt.pie(res, labels=df.columns[1:-2])
plt.show()
```
